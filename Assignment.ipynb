# Q) 5 a, Volatility

```python
google = data.DataReader('GOOGL', 'iex', start, end)
apple = data.DataReader('AAPL', 'iex', start, end) 
facebook = data.DataReader('FB','iex',start,end)
netflix = data.DataReader('NFLX','iex',start,end)
amazon = data.DataReader('AMZN','iex',start,end)
```

## volatility of google

```python
google.head(2)
```

## step 1    
  #### find  the log price change

```python
google['log_price'] = np.log(google.close/google.close.shift(1)) 
```

###   log price is added to new colunm  named  "log_price"

```python
google.head(4)
```

### Step 2  : 
 #### find the standard deviation of the log price



```python
Stnd_dvtn = google['log_price'].std()
print('standard deviation :', Stnd_dvtn)
```

### Step3:
###  standard deviantion * sqrt(250)     (250 is number of market day)

```python
volatilty = Stnd_dvtn * (math.sqrt(250))
print('volatility :', volatilty)
```

###  volatility of google is : 15.13%
